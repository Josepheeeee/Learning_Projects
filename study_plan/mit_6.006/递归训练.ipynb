```python
def factorial(n):
    # 1. 基准情况
    # 2. 递归调用与返回结果
    while n > 1 :
        return n*factorial(n-1)
    if n == 1:
        return 1
```

```python
factorial(5)
```

```python
def fibonacci(n):
    # 1. 基准情况
    # 2. 递归调用（注意：这里会有两次递归调用）
    while n >= 2:
        return fibonacci(n-1)+fibonacci(n-2)
    if n == 0 :
        return 0
    if n == 1 :
        return 1
    pass
```

```python
fibonacci(6)
```

```python
def sum_array(A, i):
    # 1. 基准情况
    # 2. 递归获取前 i 个数的和，然后加上 A[i]
    while i > 0 :
        return A[i] + sum_array(A,i-1)
    if i == 0:
        return A[0]
    pass
```

```python
A = [0,1,2,3,4,5]
sum_array(A,5)
```

```python
#反转字符不太会
def reverse_string(s):
    # 1. 基准情况
    # 2. 递归：取出最后一个字符，拼接上对前面部分的翻转结果
    if len(s) <= 1:
        return s
    
    # 递归步骤：
    # s[-1] 是最后一个字符
    # s[:-1] 是除了最后一个以外的所有字符（切片）
    return s[-1] + reverse_string(s[:-1])
    pass
```

```python
def climb_stairs(n):
    # 1. 基准情况
    # 2. 递归分解逻辑
    if n > 2:
        return climb_stairs(n-1) + climb_stairs(n-2)
    if n == 1:
        return 1
    if n == 2:
        return 2
    pass
```

```python
climb_stairs(3)
```

```python
def binary_search_recursive(A, target, low, high):
    # 1. 基准情况
    # 2. 计算 mid
    # 3. 递归：去左半边或右半边
    mid = (low + high) // 2
    if target < A[mid]:
        binary_search_recursive(A, target, low, mid) 
    elif target > A[mid]:
        binary_search_recursive(A, target, mid, high) 
    else:
        return mid
    pass
```

```python
def binary_search_recursive(A, target, low, high):
    # 基准情况：搜索范围为空
    if low > high:
        return -1
        
    mid = (low + high) // 2
    if A[mid] == target:
        return mid
    elif target < A[mid]:
        # 必须 return，否则结果传不回顶层
        return binary_search_recursive(A, target, low, mid - 1)
    else:
        return binary_search_recursive(A, target, mid + 1, high)
```

```python
print(binary_search_recursive([1,2,3,4,5,6],5,0,6))
```

```python
def fast_power(x, n):
    if n == 0: return 1
    
    # 关键：只调用一次递归，存入变量
    half = fast_power(x, n // 2)
    
    if n % 2 == 0:
        return half * half
    else:
        return x * half * half
```

```python
def fast_power(x, n):
    # 1. 基准情况
    # 2. 递归：先算一半的结果，再根据奇偶性求平方
    if n == 0:
        return 1
    if n % 2 == 0:
        return fast_power(x,n//2)*fast_power(x,n//2)
    else:
        return x * fast_power(x,n//2)*fast_power(x,n//2)
    pass
```
